### 1.1 k-fold cross validation and tuning parameters

```python
from sklearn.model_selection import KFold
def rmsle(y_true, y_pred):
    assert len(y_true) == len(y_pred)
    return np.sqrt(np.mean(np.square(np.subtract(np.log1p(y_true), np.log1p(y_pred)))))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer
kf = KFold(n_splits=5, shuffle=True, random_state=1111)
kf

model = LogisticRegression(penalty = 'l2')
param_grid = {
    'C' : [0.01,0.1,1,10,100,1000]    
}
search_cv = GridSearchCV(model, param_grid, scoring=make_scorer(rmsle, greater_is_better=False),
                   cv=kf, verbose=3)

search_cv.fit(X_train, Y_train)
```

## 4 PCA with KNN Model

### 4.1 scale trainning data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
# Fit on training set only.
scaler.fit(X_train)
# Apply transform to both the training set and the test set.
train_img = scaler.transform(X_train)
test_img = scaler.transform(X_test)
```

### 4.2 project original data into new PCs 

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=20)
pca.fit(train_img)

train_img = pca.transform(train_img)
test_img = pca.transform(test_img)
```

```python
train_img1 = pd.DataFrame(data = train_img)
test_img1 = pd.DataFrame(data = test_img)

```

### 4.3 tuning KNN model

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer
from sklearn.model_selection import KFold

param_grid = {
'n_neighbors':[8,9,10,11],
    'metric':['euclidean','manhattan']
}

kf = KFold(n_splits=5, shuffle=True, random_state=1111)
kf
model= KNeighborsClassifier()

search_cv = GridSearchCV(model, param_grid, scoring=make_scorer(rmsle, greater_is_better=False),
                   cv=kf, verbose=3)

search_cv.fit(train_img1.sample(n=50000, random_state=1), Y_train.sample(n=50000, random_state=1))


```

```python
import seaborn as sns
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

rmsle = {}
for i in (1,2,3,4,5):
    def rmsle1(y, y0):
        return np.sqrt(np.mean(np.square(np.log1p(y) - np.log1p(y0))))
    
    y_pred = search_cv.best_estimator_.predict(test_img)
    rmsle[i] = rmsle1(y_pred,Y_test)
print(np.array(list(rmsle.values())).mean())

df1 = pd.DataFrame(data=search_cv.cv_results_)
df1
fig,ax = plt.subplots()
sns.pointplot(data=df1[['mean_test_score',
                           'param_n_neighbors']],
             y='mean_test_score',x='param_n_neighbors',ax=ax)
ax.set(title="Effect of Depth on Model Performance")
results = pd.DataFrame(search_cv.cv_results_).sort_values(by='mean_test_score', ascending=False)
results.head()
```
